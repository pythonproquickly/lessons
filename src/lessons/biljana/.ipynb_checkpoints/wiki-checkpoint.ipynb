```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

```python
import json
```

```python
FORMATS = ('wiki.txt', 'exonaut.txt', 'adl.txt')
for format in FORMATS:
    json_data = json.loads(open(format).read())
    wiki_data = {}
    for key, value in json_data.items():
        if isinstance(value, dict):
            for ik1, iv1 in value.items():
                if isinstance(iv1, dict):
                    for ik2, iv2 in iv1.items():
                        if isinstance(iv2, dict):
                            for ik3, iv3 in iv2.items():
                                rows = []
                                rows.append(iv3)
                                wiki_data[f"{key}_{ik1}_{ik2}_{ik3}"] = rows
                        else:
                            rows = []
                            rows.append(iv2)
                            actual_key = f"{key}_{ik1}_{ik2}"
                            wiki_data[actual_key] = rows
                else:
                    rows = []
                    rows.append(iv1)
                    wiki_data[f"{key}_{ik1}"] = rows

        else:
            rows = []
            rows.append(value)
            wiki_data[key] = rows

    print()
    print(format)
    print("-" * len(format))
    parsed = pd.DataFrame(wiki_data)
    display(parsed)
    print()
    print(parsed.info())



```
