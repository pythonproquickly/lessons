```python
import sys
import csv

import pandas as pd

ANALYSIS_FILE = "/Users/andy/ws/lessons/data/sam/output/log_summary.csv"
la = pd.read_csv(ANALYSIS_FILE, sep='\t', engine='python', quoting=csv.QUOTE_NONE)
```

```python
la["yyyymm"] = la.time.str[:7]
la_result = la.groupby(["client", "user_agent", "yyyymm"]).agg(monthly_count = ("time", 'nunique')).unstack(fill_value=0)
# group by count by day but 1 per day for each client / agent
```
